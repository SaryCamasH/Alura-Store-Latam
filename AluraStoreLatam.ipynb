#1. Análisis de facturación



## Ingreso total por cada tienda

Calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_total_tienda1:,.2f}")
print(f"Ingreso total Tienda 2: {ingreso_total_tienda2:,.2f}")
print(f"Ingreso total Tienda 3: {ingreso_total_tienda3:,.2f}")
print(f"Ingreso total Tienda 4: {ingreso_total_tienda4:,.2f}")
```

```python
ingresos_totales = pd.Series({
    'Tienda 1': ingreso_total_tienda1,
    'Tienda 2': ingreso_total_tienda2,
    'Tienda 3': ingreso_total_tienda3,
    'Tienda 4': ingreso_total_tienda4
})

ingresos_totales_ordenados = ingresos_totales.sort_values(ascending=False)

print("Ingresos totales por tienda (ordenados de mayor a menor):\n")
print(ingresos_totales_ordenados)
print(f"\nLa tienda con mayores ingresos es: {ingresos_totales_ordenados.index[0]} con {ingresos_totales_ordenados.iloc[0]:,.2f}")
print(f"La tienda con menores ingresos es: {ingresos_totales_ordenados.index[-1]} con {ingresos_totales_ordenados.iloc[-1]:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.ticker as mticker

# Crear un DataFrame a partir de la Serie ingresos_totales_ordenados
ingresos_df = ingresos_totales_ordenados.reset_index()
ingresos_df.columns = ['Tienda', 'Ingreso Total']

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Tienda',
    y='Ingreso Total',
    data=ingresos_df,
    palette='viridis'
)
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total (COP)')

# Formatear el eje Y con comas
formatter = mticker.StrMethodFormatter('{x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

 ## Ventas por categoría

 Calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por Categoría en Tienda 1:")
print(ventas_por_categoria_tienda1)
```

```python
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por Categoría en Tienda 2:")
print(ventas_por_categoria_tienda2)
```

```python
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por Categoría en Tienda 3:")
print(ventas_por_categoria_tienda3)
```

```python
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por Categoría en Tienda 4:")
print(ventas_por_categoria_tienda4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ventas_por_categoria_todas_tiendas = pd.DataFrame({
    'Tienda 1': ventas_por_categoria_tienda1,
    'Tienda 2': ventas_por_categoria_tienda2,
    'Tienda 3': ventas_por_categoria_tienda3,
    'Tienda 4': ventas_por_categoria_tienda4
})

ventas_por_categoria_todas_tiendas = ventas_por_categoria_todas_tiendas.reset_index()

ventas_melted = ventas_por_categoria_todas_tiendas.melt(
    id_vars=['Categoría del Producto'],
    var_name='Tienda',
    value_name='Cantidad de Ventas'
)

plt.figure(figsize=(14, 8))
sns.barplot(
    data=ventas_melted,
    x='Categoría del Producto',
    y='Cantidad de Ventas',
    hue='Tienda',
    palette='viridis'
)
plt.title('Ventas por Categoría para Todas las Tiendas')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

## Productos más vendidos y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
ventas_por_producto_tienda1 = tienda['Producto'].value_counts()

top_3_productos_tienda1 = ventas_por_producto_tienda1.head(3)
bottom_3_productos_tienda1 = ventas_por_producto_tienda1.tail(3)

print("Top 3 Productos más vendidos en Tienda 1:")
print(top_3_productos_tienda1)
print("\nBottom 3 Productos menos vendidos en Tienda 1:")
print(bottom_3_productos_tienda1)
```

```python
ventas_por_producto_tienda2 = tienda2['Producto'].value_counts()

top_3_productos_tienda2 = ventas_por_producto_tienda2.head(3)
bottom_3_productos_tienda2 = ventas_por_producto_tienda2.tail(3)

print("Top 3 Productos más vendidos en Tienda 2:")
print(top_3_productos_tienda2)
print("\nBottom 3 Productos menos vendidos en Tienda 2:")
print(bottom_3_productos_tienda2)
```

```python
ventas_por_producto_tienda3 = tienda3['Producto'].value_counts()

top_3_productos_tienda3 = ventas_por_producto_tienda3.head(3)
bottom_3_productos_tienda3 = ventas_por_producto_tienda3.tail(3)

print("Top 3 Productos más vendidos en Tienda 3:")
print(top_3_productos_tienda3)
print("\nBottom 3 Productos menos vendidos en Tienda 3:")
print(bottom_3_productos_tienda3)
```

```python
ventas_por_producto_tienda4 = tienda4['Producto'].value_counts()

top_3_productos_tienda4 = ventas_por_producto_tienda4.head(3)
bottom_3_productos_tienda4 = ventas_por_producto_tienda4.tail(3)

print("Top 3 Productos más vendidos en Tienda 4:")
print(top_3_productos_tienda4)
print("\nBottom 3 Productos menos vendidos en Tienda 4:")
print(bottom_3_productos_tienda4)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_top_products(top_products_series, store_name):
    plt.figure(figsize=(10, 6))
    sns.barplot(
        x=top_products_series.index,
        y=top_products_series.values,
        palette='magma'
    )
    plt.title(f'Top 3 Productos Más Vendidos en {store_name}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad de Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Graficar el Top 3 para cada tienda
plot_top_products(top_3_productos_tienda1, 'Tienda 1')
plot_top_products(top_3_productos_tienda2, 'Tienda 2')
plot_top_products(top_3_productos_tienda3, 'Tienda 3')
plot_top_products(top_3_productos_tienda4, 'Tienda 4')
```
